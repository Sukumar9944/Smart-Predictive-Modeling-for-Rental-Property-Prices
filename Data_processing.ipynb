##### **Data Cleaning**

```python
# 1. Removing null values

new_df = df.dropna()
new_df.isnull().sum()
```

```python
# 2. Dropping dulpicated rows

drop_df = new_df.drop_duplicates()
drop_df.duplicated().sum()
```

```python
pd.set_option('display.max_columns', 25)
```

```python
drop_df.head()
```

```python
# Dropping unneccesary columns

analysis_df = drop_df.drop(['id', 'activation_date', 'amenities', 'locality'], axis = 1)
analysis_df.head()
```

##### **Detecting Outliers using Box Plot**

```python
import plotly.express as px

# 1. Property Size

fig = px.box(analysis_df['property_size'])
fig.show()
```

```python
# 2. Rent

fig = px.box(analysis_df['rent'])
fig.show()
```

```python
# 3. property_age

fig = px.box(analysis_df['property_age'])
fig.show()

'In this box plot we can see that the max age of the property is 400, It is a error'
```

```python
ml_df = analysis_df.query('property_age !=400 and property_age >= 1')

new_df = ml_df.query('cup_board >=0 and cup_board <= 40')

fig = px.box(new_df['property_age'])
fig.show()

fig = px.box(new_df['cup_board'])
fig.show()

```

##### **Getting the value counts of Categorical columns**

```python
ml_df.sample()
```

```python
new_df['type'] = new_df['type'].replace('bhk2', 'BHK2')
new_df['type'] = new_df['type'].replace('bhk3', 'BHK3')
new_df['type'] = new_df['type'].replace('1BHK1', 'BHK1')

new_df['water_supply'] = new_df['water_supply'].replace('CORP_BORE', 'CORPORATION')
```

```python
category_df = ml_df[['type','lease_type','furnishing','parking','facing','water_supply','building_type']]
category_df.head()

for i in category_df.columns:
    print(f'The value counts of {i} : {category_df[i].value_counts()}')
    print('__________________________________________________________')
```
